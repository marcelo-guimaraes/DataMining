```python
import requests
import urllib
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
from bs4 import BeautifulSoup
```

```python
page = 'https://sistemas.anm.gov.br/arrecadacao/extra/relatorios/arrecadacao_cfem.aspx'
page = requests.get(page)
soup = BeautifulSoup(page.text, 'html.parser')
soup = soup.get_text(strip=True)
```

```python
lista = soup.split(",")
lista
```

```python
Arrecadação_CFEMA = pd.DataFrame({"coluna": lista})
```

```python
Arrecadação_CFEMA.drop(0, inplace = True)
Arrecadação_CFEMA = Arrecadação_CFEMA.head(179)
```

```python
ls=[]
for x in Arrecadação_CFEMA.coluna:
    ls.append(x[2:])
    
df = pd.DataFrame({"Col":ls})    
df   
```

```python
months = []
for x in ls:
    months.append(" ".join(re.findall("[a-zA-Z]+", x)))
months = pd.Series(months).str.replace('Mar o', 'Março')    
```

```python
arrecadacao = []
for j in ls:
    arrecadacao.append(re.sub(r"[a-zA-Z]+", '', j,re.I))
arrecadacao = pd.Series(arrecadacao).str.replace('ç','')
arrec_float = arrecadacao.str.replace('.','').astype(float)
```

```python
df_arrecadacao = pd.DataFrame({"Mes":months, "Arrecadacao":arrecadacao, "Arrecadacao Float":arrec_float})
df_arrecadacao.head(5)
```

```python
df_arrecadacao.to_csv('arrecadacao.csv',index=False)
```
