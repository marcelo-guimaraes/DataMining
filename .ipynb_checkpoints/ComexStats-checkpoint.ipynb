```python
url = 'http://www.mdic.gov.br/index.php/comercio-exterior/estatisticas-de-comercio-exterior/base-de-dados-do-comercio-exterior-brasileiro-arquivos-para-download'
response = requests.get(url)
html = response.content
soup = BeautifulSoup(response.text, 'html.parser')
table = soup.find('table', class_="table table-condensed centered-tds")
links = table.find_all('a')

# links
[l.get('href') for l in links]   
clean_links = [l for l in links if l.get('href') != None]

# construindo os links
Urlinicio = 'http://www.mdic.gov.br/'
relative_urls = [link.get('href') for link in clean_links]

#EXPORTAÇÃO
table_exportacao = soup.find('table', class_="table table-condensed centered-tds")
links_exportacao = table_exportacao.find_all('a')

# links
[l.get('href') for l in links_exportacao]   
clean_links_exportacao = [l for l in links_exportacao if l.get('href') != None]

# construindo os links
Urlinicio = 'http://www.mdic.gov.br/'
urls_exportacao = [link.get('href') for link in clean_links_exportacao]

url_new_exportacao = []
for i in range(0,len(relative_urls)):
    new_element = Urlinicio + urls_exportacao[i]
    url_new_exportacao.append(new_element)
```

```python
profile = webdriver.FirefoxProfile()

# Configurando Firefox para n~ao mostrar popup quando for salvar arquivo
profile.set_preference("browser.download.panel.shown", False)
profile.set_preference("browser.helperApps.neverAsk.openFile","text/csv,application/vnd.ms-excel")
profile.set_preference("browser.helperApps.neverAsk.saveToDisk", "text/csv,application/vnd.ms-excel")
profile.set_preference("browser.download.folderList", 2)

### Mudar diretorio
profile.set_preference("browser.download.dir", "/home/marcelo/datasprints/IBRAM")

for i in relative_urls:
    # create webdriver object 
    driver = webdriver.Firefox(firefox_profile=profile) 
    url = Urlinicio + i
    
    try:
        # entra no link para baixar o arquivo
        driver.get(url) 
    except:
        # fecha a aba
        driver.close()
```

```python
def clear_products(df):
    
    import pandas as pd
    import numpy as np
    
    codigos = [72029300, 71081100, 71081210, 71081290, 71081310, 71081390, 71082000, 7112900, 68010000,
           68022100, 68022300, 68022900, 68029100, 68029200, 68029390, 68029990, 68030000]

    # Filtra os codigos de produtos estabelecidos
    df1 = df[df.CO_NCM.isin(codigos)]
    # Filtra produtos cujos codigos iniciam com 25 ou 26
    df2 = df[df.CO_NCM.astype(str).str.startswith(("25","26")) == True]
    # Junta os dados de todos os produtos que devem permanecer
    df_final = pd.concat([df1,df2])
    # Retorna o tipo da coluna CO_NCM para int
    df_final['CO_NCM'] = df_final['CO_NCM'].astype(int)
    
    
    # Apaga as coluna que nao estao no site do ComexStats
    df_final.drop(['KG_LIQUIDO','CO_PAIS','CO_UNID','CO_VIA','CO_URF'],axis=1,inplace=True)
    # Altera o a coluna de meses
    df_final['CO_MES'] = df_final.CO_MES.replace({1:"Janeiro",2:"Fevereiro",3:"Marco",4:"Abril",5:"Maio",6:"Junho",7:"Julho",
                                   8:"Agosto",9:"Setembro",10:"Outubro",11:"Novembro",12:"Dezembro"})
    # Ordeno a coluna de meses para que nao estajam em ordem alfabeta
    meses = ['Dezembro','Novembro','Outubro','Setembro','Agosto','Julho','Junho','Maio','Abril','Marco',
             'Fevereiro','Janeiro']
    df_final.CO_MES = pd.Categorical(df_final.CO_MES, categories=meses, ordered=True)
    
    # Substituo o nome das colunas para ficarem iguais ao site
    df_final.rename(columns=({"CO_ANO":"Ano","CO_MES":"Mes","CO_NCM":"Codigo NCM",
                              "QT_ESTAT":"Quant. Estatistica", "VL_FOB":"Valor FOB(US$)"}),inplace=True)
    
    # Agrupa os dados pelo mes e codigo, somando os valores FOB (US$) e Quant. Estatistica
    df_final = df_final.groupby(['Ano','Mes', 'Codigo NCM']).sum()
    
    
    # Importo dataframe com as descricoes do NCM 
    df_ncm = pd.read_csv('NCM.csv',sep=';',encoding = "ISO-8859-1")
    # Crio dicionario com os codigos e descricoes
    key = df_ncm.CO_NCM
    values = df_ncm.NO_NCM_POR
    dic = dict(zip(key,values))
    #Adiciono as descricoes no dataframe original
    df_final['Descricao NCM'] = df_final.index.get_level_values(2).map(dic)
    
    
    return df_final
```

```python
import pandas as pd
df = pd.read_csv('EXP_1997.csv',sep=';')
df = clear_products(df)
df
```
